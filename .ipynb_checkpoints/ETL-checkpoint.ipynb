```python
import win32com.client
import pandas as pd
import numpy as np

import requests

import configparser
from datetime import datetime,timedelta

import logging

import pytz

import subprocess
import os
import math
os.environ["GOOGLE_APPLICATION_CREDENTIALS"] = "cred.json"

from google.cloud import bigquery


config = configparser.ConfigParser()
config.sections()
config.read('PIMX_Config.ini')
```

```python
StdError_FileName = config['StdError']['FILE_PREFIX'] + config['FileGen']['SITEID'] + datetime.today().strftime("%Y-%m-%d-%H-%M-%S")

logger = logging.getLogger(StdError_FileName)

c_handler = logging.StreamHandler()
f_handler = logging.FileHandler(config['StdError']['OUTPUT_FOLDER'] + StdError_FileName + '.log')
c_handler.setLevel(logging.WARNING)
f_handler.setLevel(logging.ERROR)

c_format = logging.Formatter('%(name)s - %(levelname)s - %(message)s')
f_format = logging.Formatter('%(asctime)s - %(name)s - %(levelname)s - %(message)s')
c_handler.setFormatter(c_format)
f_handler.setFormatter(f_format)

logger.addHandler(c_handler)
logger.addHandler(f_handler)
```

```python
try:

    pitags_df = pd.read_csv(config['FilePath']['TAG_INPUT_FILE'])
    pitags_df.head()
except Exception as e:
    logger.error("Exception occurred", exc_info=e)
```

```python
try:
    
    column_names = config['FileGen']['SECOND_LINE'].split(config['FileGen']['DELIMETER_HEADER'])
    df_merge_pi = pd.DataFrame(columns = column_names)

    #For unexported Tags
    column_names_StdOut = config['FileGen']['SECOND_LINE'].split(config['FileGen']['DELIMETER_HEADER'])
    StdOutDf = pd.DataFrame(columns = column_names_StdOut)
    

except Exception as e:
    logger.error("Exception occurred", exc_info=e)
```

```python
try:
    pi_sdk = win32com.client.Dispatch('PISDK.PISDK')
    conn = win32com.client.Dispatch('PISDKDlg.Connections')
    pi_server = pi_sdk.Servers(config['PI_CFG']['SERVER'])
    conn.Login(pi_server,config['PI_CFG']['USER'],'',1,0) 
except Exception as e:
    logger.error("Exception occurred", exc_info=e)
```

```python
current_time = pytz.timezone("Asia/Bangkok").localize(datetime.now())
start_time = (current_time - timedelta(hours=1)).strftime("%d-%b-%y %H:10").lower()
end_time = current_time.strftime("%d-%b-%y %H").lower()
# current_time = pytz.timezone("Asia/Bangkok").localize(datetime.now())
# interval = 5
# delta_minute = int(current_time.strftime('%M')) % interval
# start_time = (current_time - timedelta(minutes=interval + delta_minute)).strftime("%d-%b-%y %H:%M").lower()
# end_time = (current_time - timedelta(minutes=delta_minute)).strftime("%d-%b-%y %H:%M").lower()
print('read data from : {} to {}'.format(start_time,end_time))
```

```python
try:
    rows_to_insert = []
#     df_merge_pi = df_merge_pi[0:0]
    for index,row in pitags_df.iterrows():
        pi_point = pi_server.PIPoints[row['PITags']] 
#         recorded_values = pi_point.Data.InterpolatedValues('26-nov-21 04:10','26-nov-21 08:00',24,"",2,None)
        recorded_values = pi_point.Data.InterpolatedValues(start_time,end_time,6,"",2,None)
        if recorded_values.Count <=0 :
            Value = np.nan
            MeasureDate=np.nan
        else:
            for raw_index,raw_data in enumerate(recorded_values) :
                Value=raw_data.Value
#                 if isinstance(Value, (int, float, complex)) != False:
#                     rows_to_insert.append({u"timestamp" : raw_data.TimeStamp.LocalDate.strftime("%Y-%m-%dT%H:%M:%S+07:00"),
#                            u"id" : row['PITags'],
#                            u"value": Value,
#                            u"created_at" : pytz.timezone("Asia/Bangkok").localize(datetime.now()).strftime("%Y-%m-%dT%H:%M:%S+07:00")
#                            })
#                 else :
#                     print('not found data !!')
                rows_to_insert.append({u"timestamp" : raw_data.TimeStamp.LocalDate.strftime("%Y-%m-%dT%H:%M:%S+07:00"),
                       u"id" : row['PITags'],
                       u"value": Value,
                       u"created_at" : pytz.timezone("Asia/Bangkok").localize(datetime.now()).strftime("%Y-%m-%dT%H:%M:%S+07:00")
                       })
#                 print(raw_index,raw_data.Value, raw_data.TimeStamp.LocalDate.strftime("%Y-%m-%dT%H:%M:%S+07:00"))
except Exception as e:
    logger.error("Exception occurred", exc_info=e)
    
# print(sorted(rows_to_insert,key=lambda x: x['timestamp']))
print('prepare data : {} rows'.format(len(rows_to_insert)))
```

```python
client = bigquery.Client()
errors = client.insert_rows_json('merlin-322103.merlin_dev.sensor_data', rows_to_insert)  # Make an API request.
if errors == []:
    print("New rows have been added.")
else:
    print("Encountered errors while inserting rows: {}".format(errors))
```
